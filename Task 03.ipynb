<h3>Import Libraries</h3>

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
import pandas as pd
```

<h3>Load the Iris dataset</h3>

```python
iris = load_iris()
```

<h3>Convert to DataFrame for easier plotting</h3>

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df["species"] = iris.target
```

<h3>Scatter Plot with seaborn</h3>

```python
plt.figure(figsize=(8,6))
sns.scatterplot(
    data=df,
    x="sepal length (cm)",
    y="petal length (cm)",
    hue="species",
    palette="deep",
    s=80
)
plt.title("Scatter Plot: Sepal Length vs Petal Length (Iris Dataset)")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Petal Length (cm)")
plt.legend(title="Species", labels=iris.target_names)
plt.savefig("Visualisation.png")
plt.show()
```

<h3>Correlation Analysis</h3>

```python
correlation = df["sepal length (cm)"].corr(df["petal length (cm)"])
print("Correlation between Sepal Length and Petal Length:", correlation)
```
